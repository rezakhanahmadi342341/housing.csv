```python
import matplotlib.pyplot as plt
from random import randint as rn
```

```python
def mx(num, _min,_max):
  point = [rn(_min,_max) for  i in range(num)]
  return point
```

```python

x = mx(5,0,100)
y = mx(5,0,100)

a = mx(5,0,100)
b = mx(5,0,100)

plt.scatter(x,y, label='x,y', alpha=0.7, c=['red','yellow','blue','black', 'pink'])
plt.scatter(a,b, label='a,b', alpha=0.7)
plt.title("scatter_plot")
plt.ylabel('scatter_y')
plt.xlabel('scatter_x')
plt.grid()
plt.legend()
plt.show()
```
