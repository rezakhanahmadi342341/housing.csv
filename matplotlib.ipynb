```python
import matplotlib.pyplot as plt
from random import randint as rn
```

```python
def mx(num, _min,_max):
  point = [rn(_min,_max) for  i in range(num)]
  return point
```

```python

x = mx(100,0,100)
y = mx(100,0,100)

a = mx(5,0,100)
b = mx(5,0,100)

plt.scatter(x,y, label='x,y',s=mx(100,10,100) ,alpha=0.7, c=mx(100,100,500), cmap='YlOrRd',
            linewidths=1.5
            )
plt.colorbar()
plt.title("scatter_plot")
plt.ylabel('scatter_y')
plt.xlabel('scatter_x')
plt.grid()
plt.legend()
plt.show()
```
